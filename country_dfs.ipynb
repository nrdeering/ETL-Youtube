```python
import csv
import pandas as pd
from sqlalchemy import create_engine

US_yt_csv = "CSV_Files/USyoutube_videos.csv"
US_yt_df = pd.read_csv(US_yt_csv)
US_yt = US_yt_df[['video_rank', 'video_id', 'country_id', 'title', 'channel_title', 'category_id', 'publish_time','views']].copy()
US_yt.head()

```

```python
FR_yt_csv = "CSV_Files/FRyoutube_videos.csv"
FR_yt_df = pd.read_csv(FR_yt_csv)
FR_yt = FR_yt_df[['video_rank', 'video_id', 'country_id', 'title', 'channel_title', 'category_id', 'publish_time','views']].copy()
FR_yt.head()
```

```python
CA_yt_csv = "CSV_Files/CAyoutube_videos.csv"
CA_yt_df = pd.read_csv(CA_yt_csv)
CA_yt = CA_yt_df[['video_rank', 'video_id', 'country_id', 'title', 'channel_title', 'category_id', 'publish_time','views']].copy()
CA_yt.head()
```

```python
rds_connection_string = "postgres:postgres@localhost:5432/countryv2_db"
engine = create_engine(f'postgresql://{rds_connection_string}')
```

```python
engine.table_names()
```

```python
US_yt.to_sql(name='US_Videos', con=engine, if_exists='append', index=False)
CA_yt.to_sql(name='CA_Videos', con=engine, if_exists='append', index=False)
FR_yt.to_sql(name='FR_Videos', con=engine, if_exists='append', index=False)
MX_yt.to_sql(name='MX_Videos', con=engine, if_exists='append', index=False)
GB_yt.to_sql(name='GB_Videos', con=engine, if_exists='append', index=False)
```
