```python
 %matplotlib notebook
import matplotlib.pyplot as plt

import pandas as pd
import json
from pandas.io.json import json_normalize
import csv
```

# CA

```python
CA_yt_csv = "CAyoutube_videos.csv"
CA_yt_df = pd.read_csv(CA_yt_csv)
CA_yt = CA_yt_df[['video_rank', 'video_id', 'country_id', 'title', 'channel_title', 'category_id', 'publish_time','views']].copy()
CA_yt.head()
```

```python
with open('json_files/CA_category_id.json') as f:
    d = json.load(f)
    
CA_cat = json_normalize(d['items'])
CA_cat['id'] = CA_cat['id'].astype(int)
CA_cat = CA_cat[['id', 'snippet.title']]
CA_cat.columns = ['category_id', 'category_title']
CA_cat
```

```python
CA_data = pd.merge(CA_yt, CA_cat, on='category_id', how='left')
CA_data
```

# US

```python
US_yt_csv = "USyoutube_videos.csv"
US_yt_df = pd.read_csv(US_yt_csv)
US_yt = US_yt_df[['video_rank', 'video_id', 'country_id', 'title', 'channel_title', 'category_id', 'publish_time','views']].copy()
US_yt.head()
```

```python
with open('json_files/US_category_id.json') as f:
    d = json.load(f)
    
US_cat = json_normalize(d['items'])
US_cat['id'] = US_cat['id'].astype(int)
US_cat = US_cat[['id', 'snippet.title']]
US_cat.columns = ['category_id', 'category_title']
US_cat
```

```python
US_data = pd.merge(US_yt, US_cat, on='category_id', how='left')
US_data
```

```python
# Create a group based on the values in the 'category_title' column
USgroup = US_data.groupby('category_title')

# Count how many times each maker appears in our group
UScount = USgroup['category_title'].count().reset_index(name='US')

UScount
```

```python
MXgroup = MX_data.groupby('category_title')
MXcount = MXgroup['category_title'].count().reset_index(name='MX')
MXcount
```

```python
FRgroup = FR_data.groupby('category_title')
FRcount = FRgroup['category_title'].count().reset_index(name='FR')
FRcount
```

```python
CAgroup = CA_data.groupby('category_title')
CAcount = CAgroup['category_title'].count().reset_index(name='CA')
CAcount
```

```python
GBgroup = GB_data.groupby('category_title')
GBcount = GBgroup['category_title'].count().reset_index(name='GB')
GBcount
```

```python
from functools import reduce

dfs = [CAcount, FRcount, GBcount, MXcount, UScount]
df_final = reduce(lambda left,right: pd.merge(left,right,on='category_title',how='outer'), dfs)
df_final
```

```python

df_final.plot(x="category_title", y=["CA", "FR", "GB", "MX", "US"], kind="bar")

plt.xlabel("Category Title")
plt.ylabel("Number of Category Title")


plt.show()
plt.tight_layout()
```
