```python
import pandas as pd
import json
from pandas.io.json import json_normalize
import csv
```

# CA

```python
CA_yt_csv = "CAyoutube_videos.csv"
CA_yt_df = pd.read_csv(CA_yt_csv)
CA_yt = CA_yt_df[['video_rank', 'video_id', 'country_id', 'title', 'channel_title', 'category_id', 'publish_time','views']].copy()
CA_yt.head()
```

```python
with open('json_files/CA_category_id.json') as f:
    d = json.load(f)
    
CA_cat = json_normalize(d['items'])
CA_cat['id'] = CA_cat['id'].astype(int)
CA_cat = CA_cat[['id', 'snippet.title']]
CA_cat.columns = ['category_id', 'category_title']
CA_cat
```

```python
CA_data = pd.merge(CA_yt, CA_cat, on='category_id', how='left')
CA_data
```
